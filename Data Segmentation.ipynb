```python
import pandas as pd
import numpy as np
```

# Load data

```python
data = pd.read_excel('Combined 2006-2021.xlsx')
print(data.shape)
data.head()
```

# Drop irrelevant features

```python
data = data.drop(["Series", "Prev Close", "VWAP", "Deliverable Volume"], axis=1)
data.head()
```

# Training dataset

```python
df_2006_to_2019 = data[data["Date"] < '2020-1-1 00:00:00'].reset_index(drop=True)
df_2006_to_2019.head()
```

# Testing dataset

```python
df_2020_to_2021 = data[data["Date"] >= '2020-1-1 00:00:00'].reset_index(drop=True)
df_2020_to_2021.head()
```

# Save datasets in csv file

```python
df_2006_to_2019.to_csv("2006-2019.csv", index=False)
df_2020_to_2021.to_csv("2020-2021.csv", index=False)
```
